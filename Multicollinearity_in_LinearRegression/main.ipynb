```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import statsmodels.api as sm
df_adv = pd.read_csv('Advertising.csv')
X = df_adv[['TV', 'radio', 'newspaper']]
y = df_adv['sales']
```

```python
df_adv.head()
```

```python
#OLS - Ordinary Least Squares
# fit a OLS model with intercept on TV and radio
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
X.iloc[:, 1:].corr()
```

```python
#The above example does not have collinearity
```

```python
# Example with collinearity
df_salary = pd.read_csv('Salary_Data.csv')
df_salary.head()
```

```python
X = df_salary[['YearsExperience', 'Age']]
y = df_salary['Salary']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```

```python
X.iloc[:, 1:].corr()
```
